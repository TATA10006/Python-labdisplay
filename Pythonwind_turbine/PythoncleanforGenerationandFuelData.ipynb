```python
# Issues arise because of the encoding, using latin1 because of the 0x1f in the error showing

GenerationandFuelData=pd.read_csv('./GenerationandFuelData.csv',engine='python',encoding='latin1')

GenerationandFuelData.isnull().sum().sort_values(ascending=False)
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

## Minimize operators by 

GenerationandFuelData['Operator Name'].value_counts()[GenerationandFuelData['Operator Name'].value_counts()>10]
```

```python
# Filtering Operator Name where the counts bigger than 10 by using below codes(where K = x, (get all items which have more than x readings) )
# https://stackoverflow.com/questions/48628417/how-to-select-rows-in-pandas-dataframe-where-value-appears-more-than-once
v = GenerationandFuelData['Operator Name'].value_counts()
Cleaneddata2=GenerationandFuelData[GenerationandFuelData['Operator Name'].isin(v.index[v.gt(9)])]
```
